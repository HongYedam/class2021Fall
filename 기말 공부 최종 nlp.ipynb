```python
type(text.split())
```

```python
words = text.split()
' '.join(text.split())
# =   ' '.join(words) 
# space로 join해라 ! (split된) words를!
```

```python
from nltk.tokenize import word_tokenize
# split도 tokenize의 원시적 함수이긴 하지만,
# 좀더 전문적인 nltk 패키지 안의 tokenize(표식화)속의 word_tokenize(단어 표식화하기=punctuation까지 다 끊어서 배열)함수를 불러옴!
# 그니까 기호랑 단어별로 
nltk.download('punkt')
# nltk. 안에 있는 다운로드 함수로, 'punkt'를 다운받아라 ->nltk.tokenize.word_tokenize는 punctuation도 다 tokenize함
# -> 그래서 이 프로그램 자체가 뭘 punkt라고 부르는 지 알아야함.
##  따라서! ### nltk에서 word_tokenize할 때에는 punkt 다운 받고 tokenize 실행해야함 !!!!

words = word_tokenize(text)
words
# 아까 text.split()은 쪼개진 단어에 punctuation도 붙어있었는데/ 
# word_tokenize는 기호는 기호별로, 단어는 단어별로 나눔=> punctuaion도 다 하나의 list로 쳐서 따로 나눠줌 
```

```python
from nltk.tokenize import RegexpTokenizer
# RegexpTokenizer = nltk안에 있는 다른 tokenize 
# = regular expression(어떤 규칙을 통과하는 것들만 찾아라)를 기반으로 한 tokenize
retokenize = RegexpTokenizer("[\w]+")
# 그 regular expression=rule에 해당하는 것들을 ()에 적어서 retokenize(이름이라 아무말해도됨) 라는 tool로 만든 것
# [\w]+ : 이말 자체가 word임!!! = 알파벳과 숫자가 하나 이상 나오는 것을 찾아라!
# 그니까 word (기호 없이)만 찾는게 규칙!!!!!
words = retokenize.tokenize(text)  # 여따가 text 대입하면 word(punctuation 생략)만으로 분리해줌
words
```

### Normalization  = 여러가지 변이형들이 있음
Stemming: am → am, the going → the go, having → hav  
(활용형 부분=어간?을 잘라버리기, normalize하기) ex) PorterStemmer/ LancasterStemmer


Lemmatization: am → be, the going → the going, having → have

(활용형을 원형(=사전에 등지된 단어)로 대체, normalize하기)
각각 여러 변이형들을 하나의 형태로 normalize하는 것.
ex) WordNetLemmatizer

```python
"PorterStemmer 쓰는 법" '있는 자체에서 잘라 주는 것'

from nltk.stem import PorterStemmer 
# 함수 PorterStemmer 과정을 stemmer로 이름 지어주고 tool로 사용하는거! 그냥 ()로 쓰면됨!
stemmer = PorterStemmer()

# stemmer.stem(): list comprehension
#                 = 위에서 tokenize해서 쪼갠 단어들(=words)를 stem하기! -> 제일 첫번째 단어만 stem함!
# ex>  stemmer.stem('having') = have나옴

[stemmer.stem(w) for w in words]
# words에서 쭉 있는 list 개수만큼 제일 첫번째부터 차례대로 for루프를 돌릴건데, 그걸 w로 출력해라!
# -> 그 한 단어, 기호씩을 차례대로 stemmer.stem(w) (=그냥 er.동사로 만듦/스테머의 stem함수)으로 넣어서 stem해라!
# 그리고 그 stem한 차례대로의 list들을 list안에 나열하라!
```

```python
"LancasterStemmer 쓰는 법"  '원형까지 분석해주는 것'

from nltk.stem import LancasterStemmer
stemmer = LancasterStemmer()
[stemmer.stem(w) for w in words]
```

```python
"WordNetLemmatizer 쓰는 법"

from nltk.stem import WordNetLemmatizer
nltk.download('wordnet') # 얘도 꼭 WordNet을 먼저 다운 받고 써야 함

lemmatizer = WordNetLemmatizer()

[lemmatizer.lemmatize(w) for w in words]
```

### Collocation, Concordance = 둘다 tokenize된 값만 입력 가능
**식= nltk.Text(words).collocations/concordance()**

두 단어가 같이 위치하는 것. 빈번하게 나오는 단어. 두 단어의 연쇄가 cowork하는 게 있다면 그걸 보는 것을 collocation이라고 함

concordance는 큰 텍스트 속에 father이라는 단어가 있다고 치면, 그 단어가 나오는 문장의 예시가 어떻게 되는가.
예문 찾을때 좋은 것!

```python
nltk.download('gutenberg') # nltk 속에 있는 corpus에 gutenberg라는 corpus가 입력돼있음->nltk 속 download 이용해서 다운받기

text = nltk.corpus.gutenberg.raw('melville-moby_dick.txt') 
# text = 모비딕의 원문 쭉 나옴
words = nltk.word_tokenize(text)      # -> 이건 punctuation, words 같이 분류돼서 나옴!
words

# '요롷게 retokenize로 해도됨!'(이미 위에 retokenize를 tool화 했으니까 바로 tool사용가능)-> puntuation 없는 words만 나옴!
# text = nltk.corpus.gutenberg.raw('austen-emma.txt')
# words = retokenize.tokenize(text)
```

```python
nltk.Text(words).collocations()
# collocation은 tokenize된 입력값만 받음!!!
# tokenize된 모비딕 words list를 입력으로 받아서(=불러와서) 
# 그것의 collocation(=함께쓰이는,cowork하는,빈번하게 사용되는 단어들)을 찾아라!
```

```python
nltk.Text(words).concordance('Great', 79, 10)
# word_tokenize된 words의 list를 불러와서, 그것을 concordance를 찾는 것!
#'Great'가 들어간 예문,용법,용매들을 찾는 것!
# 79 = Great를 중심으로했을때, Great를 포함해서 앞뒤로 나오는 character(띄어쓰기 포함)의 개수-1개
# -> 그니까 Great를 중심으로한 78개의 character을 가져와라! 앞의 다른 문장이어도 연결해서 가져와야함!
# 10 = 예문의 개수 ! / 10개 보여줘라!
```

```python
nltk.Text(words).dispersion_plot(["Emma", "Knightley", "Frank", "Jane", "Harriet", "Robert"])
# dispersion_plot= 이 단어들이 words 속에서 얼마나 퍼져있는가. -> 플라팅해라 !
```

```python
nltk.Text(words).similar("Emma")
# similar함수= Emma랑 similar하게 쓰이는 것들?
```

```python
nltk.Text(words).common_contexts(["Emma", "she"])
# Emma하고 She라는 단어랑 같이 쓰이는 양옆 문장성분/문맥/내용들을 분석해라!
#_가 엠마나 쉬가 들어갔던 자리
```

### Regular expression (정규표현= 상상하는 규칙을 regular하게 표현하는 법) = 입력값은 string!

자연어처리=텍스트처리=natural language processing (nlp)처리할 때 제일 기본이 되는 스킬! 어떤 corpus가 있으면 어떤 단어에서 철자 character가 있으면, 두번째 character가 r이고 5번째 character이 만약 있다면, t인 것들을 모두 찾아내라! / 모음이 2개 혹은 3개가 있는 단어를 다 찾아내라! 


-이렇게 내가 원하는 규칙을 전체 corpose에 던질 수 있음!
이때 생각하는 규칙을 텍스트에 적용시켜서 뭔가의 결과를 가지고 올 때, 이를 regular expression이라고 함!

```python
import re

# re안에는 search 함수, findall 함수, replace함수,
```

```python
'''       Basic Regular Expression Meta-Characters, Including Wildcards, Ranges and Closures

.	        Wildcard, matches any character
^abc	    Matches some pattern abc at the start of a string
abc$	    Matches some pattern abc at the end of a string
[abc]	    Matches one of a set of characters
[^abc]    Matches anything but a set of characters
[A-Z,0-9]	Matches one of a range of characters
ed|ing|s	Matches one of the specified strings (disjunction)
*	        Zero or more of previous item, e.g. a*, [a-z]* (also known as Kleene Closure)
+	        One or more of previous item, e.g. a+, [a-z]+
?	        Zero or one of the previous item (i.e. optional), e.g. a?, [a-z]?
{n}	      Exactly n repeats where n is a non-negative integer
{n,}	    At least n repeats
{,n}	    No more than n repeats
{m,n}	    At least m and no more than n repeats
a(b|c)+	  Parentheses that indicate the scope of the operators
(...)     Matches whatever regular expression is inside the parentheses
\d
Matches any decimal digit; this is equivalent to the class [0-9].
\D
Matches any non-digit character; this is equivalent to the class [^0-9].
\s
Matches any whitespace character; this is equivalent to the class [ \t\n\r\f\v].
\S
Matches any non-whitespace character; this is equivalent to the class [^ \t\n\r\f\v].
\w
Matches any alphanumeric character; this is equivalent to the class [a-zA-Z0-9_].

\W
Matches any non-alphanumeric character; this is equivalent to the class [^a-zA-Z0-9_].

'''
```

```python
re.search('ed', 'considered')
# 앞부분이 규칙 , 대상이 'considered'라는 string! 
# search = 이 'considered'라는 string에서, ed의 index 순서를 찾아라!

re.findall('ed', 'considered seed')
# findall = 출력값이 []안에 그대로 ed로 나옴. => ['ed', 'ed']

re.findall('ed$', 'considered seed')
# 앞 부분이 'ed$'로 달러가 붙으면, 제일 마지막거 하나만 찾아라! => ['ed']

re.findall('^con', 'considered condo')
# '^con'처럼 ^ 꺽쇠표시가 re 앞에 붙으면, string이 시작하는 곳 innitial에 있는 것을 찾으라는 말임!
# => ['con'] (considered에서 나온 con임!)

re.findall('[cd]', 'considered condo') # c 혹은 d     =>['c', 'd', 'd', 'c', 'd'] 

re.findall('[^abc]', 'considered condo')  # a도 아니고 b도 아니고 c도 아니고 (띄어쓰기도 나옴)
# => ['o', 'n', 's', 'i', 'd', 'e', 'r', 'e', 'd', ' ', 'o', 'n', 'd', 'o']

re.findall('[a-z]', 'considered condo')  #a에서 z까지 중에 있는 철자들 다 찾아라(a-z역슬래쉬s하면 space도 같이 찾아줌)
# =>['c', 'o', 'n', 's', 'i', 'd', 'e', 'r', 'e', 'd', 'c', 'o', 'n', 'd', 'o']

 #'[^a-z]': a-z가 아닌 철자들을 다 찾아라! => 대문자나 space, punctuation, 숫자 다 나옴

re.findall('ed|ing|s', 'considered condo')  #ed혹은 ing혹은 s를 찾아라!
# => ['s', 'ed']

re.findall('[s]*', 'consssidered condo') # s가 0개 이상 들어간걸 찾아라 => 0개도 포함되는 것임
# => ['', '', '', 'sss', '', '', '', '', '', '', '', '', '', '', '', '', '']
# [s]+ : s가 하나 이상
# [s]? : s가 없거나 하나 일 때
# [s]{2} : s가 딱 두번 나오는 것
# [s]{1,} : s가 1번 이상 나오는 것
# [s]{,1} : s가 1번 이하 나오는 것
# [s]{1,4} : s가 1번부터 4번까지 반복돼서 나오는 것
# [s]\d : s가 .................?
```

```python
# 합쳐지는 다른 예시
word = 'supercalifragilisticexpialidocious'
re.search('[aeiou]li',word)
# 위에서 가져온 regular expression의 search라는 함수를 쓰는 것임
# Argument 2개 있음-> 괄호 첫번째 = regular expresion=규칙을 두번째 = word에서 찾아라 !

# [abc]	    Matches one of a set of characters - 이 중에 하나가 포함되어있으면 찾아라!
# [aeiou]li = ali / eli / ili / oli / uli 를 찾아라 -> 이중에 제일 첫번째 나오는 걸 찾아주면 됨

# 답 span=(6, 9) 뜻 s를 첫번째로봐서, 7번째부터 9번째까지/ 

# .li = .자리에는 뭐든 와도 상관없다!
```

### Extract information (pos tag, named entity)



```python
sent = "I am Jhon from America and would like to go to Starbucks"
#이런 raw text= 문장 string 하나 주어져있다고 생각하자.
# 이게 뭐가 명사구고, 전치사고, 하는 품사 pos(=part of speech) tagging(이름붙이기)를 자동으로 해준다. 
# named entity는, Pos Tag가 주어져 있을때, 조직인지, 사람인지, 위치, 날짜, 시간, 돈, 등의 카테고리에 포함시켜주는 것.
words = nltk.word_tokenize(sent)
```

```python
'''
POS tag list:

CC	coordinating conjunction - 접속사
CD	cardinal digit
DT	determiner -관사
EX	existential there (like: "there is" ... think of it like "there exists")
FW	foreign word
IN	preposition/subordinating conjunction (전치사)
JJ	adjective	'big'
JJR	adjective, comparative	'bigger'
JJS	adjective, superlative	'biggest'
LS	list marker	1)
MD	modal	could, will
NN	noun, singular 'desk' -명사
NNS	noun plural	'desks'
NNP	proper noun, singular	'Harrison'
NNPS	proper noun, plural	'Americans'
PDT	predeterminer	'all the kids'
POS	possessive ending	parent's
PRP	personal pronoun	I, he, she
PRP$	possessive pronoun	my, his, hers
RB	adverb	very, silently,
RBR	adverb, comparative	better
RBS	adverb, superlative	best
RP	particle	give up
TO	to	go 'to' the store.
UH	interjection	errrrrrrrm
VB	verb, base form	take - 동사
VBD	verb, past tense	took
VBG	verb, gerund/present participle	taking
VBN	verb, past participle	taken
VBP	verb, sing. present, non-3d	take (현재 동사)
VBZ	verb, 3rd person sing. present	takes
WDT	wh-determiner	which
WP	wh-pronoun	who, what
WP$	possessive wh-pronoun	whose
WRB	wh-abverb	where, when
'''
```

```python
nltk.download('averaged_perceptron_tagger')
# pos_tagging을 하기 위해서 nltk에 있는pos(품사)tagger(=averaged_perceptron_tagger) 다운받고,
pos = nltk.pos_tag(words)    # 위에서 tokenize해준 words를 pos tagging 해주기 
pos
```

```python
nltk.download('maxent_ne_chunker')   # nltk에 있는 maxent_ne_chunker를 또 다운로드하기
NE = nltk.ne_chunk(pos)   # NE = named entity
# common Entity types: ORGANIZATION, PERSON, LOCATION, DATE, TIME, MONEY, and GPE (geo-political entity= 위치, 장소)
print(NE) # 7가지의 common한 entity type으로, 중요한 정보를 물어봄
```

### Wordcloud = 단어 시각화



```python
from wordcloud import WordCloud, STOPWORDS  # wordcloud 패키지 불러오기 # STOPWORDS = wordcloud에서 제공하는 stopwords
import matplotlib.pyplot as plt

text = nltk.corpus.gutenberg.raw('bible-kjv.txt')   # nltk에서 제공하는 텍스트 불러오기

wc = WordCloud().generate(text) # 어떤 단어 시각화 이미지의 출력값이 생성되고 (이때 괄호에 stopword 쓸 수 있는데 안씀!)
plt.imshow(wc)   # 그걸 matplotlib.pyplot에 들어있는 plot function(image show)을 써서 보이게 만들어줌
```

```python
# 근데 어떤 단어들은 너무 많이 쓰여서, 그건 제외하고싶다! -> stopwords 사용
# 


stopwords = set(STOPWORDS) # wordcloud에서 제공하는 stopwords(STOPWORDS) = stopwords의 집합set이다! 지정해주기
stopwords.add('unto')  # stopwords에 어떤 단어('unto'같은 단어)들을 추가해서 또 빼고 싶다! 하면 stopwords.add()하면 됨!

wc = WordCloud(stopwords = stopwords).generate(text) 
plt.imshow(wc)
```
